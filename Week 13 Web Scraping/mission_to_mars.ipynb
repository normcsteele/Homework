````python
mars_html  = requests.get("https://space-facts.com/mars/")
soup = bs(mars_html.text, 'html.parser')
print(soup)
````

````python
table_scraper = soup.find_all("table", id="tablepress-mars")[0]
table_list = []

for x in table_scraper.find_all("tr"):
    entry = x.text.strip().split(":")
    table_list.append({
        "Description": entry[0],
        "Value": entry[1]
        })

table_df = pd.DataFrame(table_list)
table_df
````
